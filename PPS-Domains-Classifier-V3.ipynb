# 1. Overview of Project <a class="anchor" id="overview-project"></a>

PPS currently uses the **Google Detection** API in order to find uses of particular pictures through out the web. Once Google Detection has found the use cases of the image, domain data is gathered, and placed in a mongodb. Based on certain input features, a rule based scoring system is in place to determine if the domain is with pursuing to prosecute. 

##  Current Rule Based Scoring
* **`essential.score.latest.score`**: the output of the scoring system (from 1 - 13, where 13 means the domain is valuable)
*  **`essential.BellonaStatus`**: first thing they go off of. Values of `SKIP_` or `HOLD_` are given scores of zero since they are not valuable. 
* **`ssential.score.latest.components.whoisScoreComponents.registrantContactCountry`**: USA gets a score of 3, some european countries, singapore, canada and austraila are given a score of 1, everything else is 0
* **`essential.score.latest.components.privateRegistrationStatus`**:  If the domain is registered privately (meaning done by a domain service, that means we can't easily determine who the owner is and thus it makes it harder prosecute, so we give these a score of 0, everything else is whatever the whois score was (which isn't zero since we already filtered those out). 
* **`essential.score.latest.components.alexaData.trafficData.ran`**:  finally, you can get a boost from alexa based on if the site is popular. Anything below 10000 gets a bonus of 10 (so for a USA site, this gives 13), and that goes down to a bonus of 2 for a score below 500000, with scores in between. 

##  Idea
> Implement a supervised machine learning classification algorithm in order to better predict what domains are worth pursuing for copy right infringement

# 2. Overview of Machine Learning <a class="anchor" id="overview-ml"></a>
Here are a few terms that are used pretty frequently and would help to have a base understanding of throughout the rest of this report. 

## Input Features
* The inputs variables that the model will use to make predictions with. Ex: **`domainAge`**

## Class Labels/Output/Response
* This refers to the class or label that is associated with a set of input features. In our case it is: **`analystResult`**.

## Training example
* A training example is made up of input variables and an output label. A machine learning model will try and find a relationship (map) between the input features and the output label. 

## Training
* The process of running a machine learning model on training data, during which the model's parameters will "learn" the relationship that maps the inputs to the outputs.

## Predicting
* Once a model has been trained, it can then take a new example(an **unseen** set of input features) and make a prediction of what class it will be belong to. If we know the class ahead of time, we can measure prediction accuracy as well as other performance metrics.

## Train/Test split (Cross Validation) - Preventing overfitting 
* Because most machine learning models will learn their parameters based off of the data they are trained on, I want to make every effort to prevent overfitting (Seen below). 



* Overfitting will occur when the model begins to tune its parameters based off of noise in the training data. This allows for the accuracy for predictions made on training data (that the model has seen before) to be very high, but often means that when new unseen test data is presented, the accuracy plummets. 
* To prevent this, I will often split my data from the start into a training and testing set.
* For example, say I have 1000 labeled examples. I may split them into a training set of 800 examples and a test set of 200 examples. I would train my model with the set of 800, and then make predictions with the 200 test examples. Because the model had not seen the 200 test examples, they are a good indicator of how it may perform

## Accuracy vs. Precision vs. Recall vs. F1-Score

There are several metrics that will be used in scoring model prediction performance in this report. But first to motivate the need for these different metrics, imagine that a model has the main goal to maximize its accuracy. Now consider the following situation: the model is trying to predict which patients may have cancer. Overall, 99% of the population is cancer free, and only 1% has the disease. Now, if the model only tries to maximize accuracy, it may just predict that **no one** has cancer, and hence it would have a 99% accuracy. Therefore we need to consider a confusion matrix in order to better optimize our results and prevent the situation above from occuring. 



True positive and true negatives are the observations that are correctly predicted and therefore shown in green. We want to minimize false positives and false negatives so they are shown in red color. These terms are a bit confusing. So let’s take each term one by one and understand it fully.

**True Positives (TP)** - These are the correctly predicted positive values which means that the value of actual class is yes and the value of predicted class is also yes. E.g. if actual class value indicates that domains is valuable and predicted class tells you the same thing.

**True Negatives (TN)** - These are the correctly predicted negative values which means that the value of actual class is no and value of predicted class is also no. E.g. if actual class says this domain is not valuable and predicted class tells you the same thing.

False positives and false negatives, these values occur when your actual class contradicts with the predicted class.

**False Positives (FP)** – When actual class is no and predicted class is yes. E.g. if actual class says this domain is not valuable but predicted class tells you that domain is valuable. Note, this is also refered to as **Type I error**.

**False Negatives (FN)** – When actual class is yes but predicted class in no. E.g. if actual class value indicates that this domain is valuable and predicted class tells you that domain is not valuable. Note, this is also refered to as **Type II error**.

Once you understand these four parameters then we can calculate Accuracy, Precision, Recall and F1 score.

### Accuracy
Accuracy is the most intuitive performance measure and it is simply a ratio of correctly predicted observation to the total observations. One may think that, if we have high accuracy then our model is best. Yes, accuracy is a great measure but only when you have symmetric datasets where values of false positive and false negatives are almost same. Therefore, you have to look at other parameters to evaluate the performance of your model.

### $$Accuracy = \frac{TP+TN}{TP+FP+FN+TN}$$

### Precision 
"How often does the algorithm cause a false alarm?". Precision is the ratio of correctly predicted positive observations to the total predicted positive observations. The question that this metric answer is of all domains that our model labeled as worth pursuing, how many actually were worth pursuing? 

### $$Precision = \frac{TP}{TP+FP}$$

### Recall
"How sensitive is our algorithm?". Recall is the ratio of correctly predicted positive observations to the all observations in actual class - yes. The question recall answers is: Of all the domains that were actually marked as worth pursuing, how many did the model classify as worth pursuing? 

### $$Recall = \frac{TP}{TP+FN}$$

### F1 Score
F1 Score is the weighted average of Precision and Recall. Therefore, this score takes both false positives and false negatives into account. Intuitively it is not as easy to understand as accuracy, but F1 is usually more useful than accuracy, especially if you have an uneven class distribution. Accuracy works best if false positives and false negatives have similar cost. If the cost of false positives and false negatives are very different, it’s better to look at both Precision and Recall.er

### $$F1 = \frac{2*(Recall * Precision)}{(Recall + Precision)}$$

I am going to be using these metrics throughout this report.

## 3.2.1 Registrant Contact Country 
* This is a categorical input (1 of many countries), and for this analysis we will look at whether the United States being the input is a statistically significant predictor of analyst result.
* To complete this test a contingency table must be created and a chi square test run on the data

```python
from scipy import stats
from scipy.stats import chi2, chi2_contingency

# contingency table
#           valuable   not valuable
#------------------------------
# USA     |   a            b
# Not USA |   c            d


USA_domains = analystScoredDomains[analystScoredDomains['registrantContactCountry'] == 'UNITED STATES']['analystResult']
non_USA_domains = analystScoredDomains[analystScoredDomains['registrantContactCountry'] != 'UNITED STATES']['analystResult']

USA_valuable = USA_domains.sum()
USA_nonvaluable = USA_domains.size - USA_domains.sum()

non_USA_valuable = non_USA_domains.sum()
non_USA_nonvaluable = non_USA_domains.size - non_USA_domains.sum()

T = np.array([[USA_valuable, USA_nonvaluable], [non_USA_valuable, non_USA_nonvaluable]])

chi2_contingency(T, correction=False)[1]
```

## Result
The final result for the p value regarding registrant contact country and its statistical significance in predicting analyst result is seen above: **7.3923077882576635e-10**.

This means that if the registrant contact country being USA was **not** a statistically significant predictor of **analystResult**, then the values in the chi contingency table would only be seen **7.3923077882576635e-8**% of the time. 

So, registrant contact country will be used a predictor. 

There are two different scenarios where registrant contact country can be utilized:
> 
1. It can be used as a binary input feature where if the value is USA it is a 1, and if it is not USA it is a 0. 
2. Through the use of one-hot encoding, each country can get its own input vector, and for each training example whatever country it is associated with, that countries vector will get a 1, and all others will be given 0's.

Both of these examples are valid, and what makes the most sense is to try and each and see how it effects our prediction metrics.

## 3.2.2 Domain Age
* This is a continuous input, and a t test will be conducted to determine its statistical significance. 

```python
analystScoredDomains['domainAge'] = analystScoredDomains['domainAge'].fillna(analystScoredDomains['domainAge'].mean())
domainAgeValuable = analystScoredDomains[analystScoredDomains['analystResult'] == 1]['domainAge']
domainAgeNotValuable = analystScoredDomains[analystScoredDomains['analystResult'] == 0]['domainAge']

tDomainAge, pDomainAge = stats.ttest_ind(domainAgeValuable,domainAgeNotValuable)
print("t statistic:", tDomainAge,"\tp value:", pDomainAge)
```

## Result
The final result for the p value concerning domain age was 0.02%. This is well under the significance value of 5% that is widely accepted. Domain age will be included in the final model as an input feature.

---
## 3.2.3 Privately Registered
* This is a binary input, and will be determined with a chi square test. 

```python
private_domains = analystScoredDomains[analystScoredDomains['privateRegistrationStatus'] == 1]['analystResult']
public_domains = analystScoredDomains[analystScoredDomains['privateRegistrationStatus'] == 0]['analystResult']

private_valuable = private_domains.sum()
private_nonvaluable = private_domains.size - private_domains.sum()

public_valuable = public_domains.sum()
public_nonvaluable = public_domains.size - public_domains.sum()

T = np.array([[private_valuable, private_nonvaluable], [public_valuable, public_nonvaluable]])

chi2_contingency(T, correction=False)[1]
```

## Result
The final result for the p value concerning private registration status as a predictor is 7.08%. This is **not** statisically significant. This was slightly expected after the visualization seen earlier. Private registration status will be taken out as a predictor initially. 

A note: It is most definitely possible that private registration status is a useful predictor in the case of the population, however, concerning this sample dataset, it was not deemed significant. If this does not match with intuition then a better sampling process will need to be designed, and the analysis repeated.

## 3.2.4 Traffic Data Rank 
* This is a continuous input, and will utilize a t-test.

```python
trafficDataRankDomains = analystScoredDomains[['analystResult', 'trafficDataRank']].dropna()

trafficValuable = trafficDataRankDomains[trafficDataRankDomains['analystResult'] == 1]['trafficDataRank'] 
trafficNotValuable = trafficDataRankDomains[trafficDataRankDomains['analystResult'] == 0]['trafficDataRank'] 

t, p = stats.ttest_ind(trafficValuable,trafficNotValuable)
print("t statistic:", t,"\tp value:", p)
```

### Metric Function

```python
def get_metrics(cm):
    accuracy = (cm[0,0]+cm[1,1])/(cm[0,0]+cm[0,1]+cm[1,0]+cm[1,1])
    precision = cm[0,0]/(cm[0,0]+cm[1,0])
    recall = cm[0,0]/(cm[0,0]+cm[0,1])
    print('Accuracy: ', accuracy, '\nPrecision:', precision, '\nRecall:', recall)
```

### Metric Imports 

```python
from sklearn.utils import shuffle
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from imblearn.over_sampling import SMOTE
```

# 3.3.1 Model A
* For model A there will be two input features:
    1. domain age
    2. USA registrant contact country binary

* And the model will be logistic regression

Now a training, cross validating, and test set will be created. 
### Train, Validate, Test set creation

```python
model_A_Domains = analystScoredDomains.copy()
model_A_Domains = shuffle(model_A_Domains)
model_A_Domains = model_A_Domains.drop(['trafficDataRank', 'privateRegistrationStatus', 'domain'], axis=1)
model_A_Domains['registrantContactCountry'] = model_A_Domains['registrantContactCountry'].map(lambda x: 1 if x == 'UNITED STATES' else 0)

train, validate, test = np.split(model_A_Domains.sample(frac=1), [int(.6*len(model_A_Domains)), int(.8*len(model_A_Domains))])

xtrain = train[['domainAge', 'registrantContactCountry']]
ytrain = train['analystResult']

xvalidate = validate[['domainAge', 'registrantContactCountry']]
yvalidate = validate['analystResult']

xtest = test[['domainAge', 'registrantContactCountry']]
ytest = test['analystResult']

# Perform SMOTE oversampling on training set
sm = SMOTE(random_state=12, ratio = 1.0)
x_train, y_train = sm.fit_sample(xtrain, ytrain)
```

## Instantiate and Train model

```python
from sklearn.linear_model import LogisticRegression
model_A = LogisticRegression(class_weight={0: 1.1, 1: 1})
model_A.fit(x_train, y_train)
```

## Make Predictions on Validate Set

```python
predictions = model_A.predict(xvalidate)
cm = confusion_matrix(yvalidate, predictions, labels=[1, 0])
get_metrics(cm)
```

## Make Predictions on Test Set

```python
predictions = model_A.predict(xtest)
cm = confusion_matrix(ytest, predictions, labels=[1, 0])
get_metrics(cm)
```

# Analysis <a class="anchor" id="analysis"></a>
Overall the above results are rather interesting. The main quantitative conclusions reached are as follows:
## 1) Domain Age and Contact Country are the only very statistically significant predictors
It was shown that domain age and registrant contact country are the only predictors thats are statistically significant in predicting whether a domain is valuable or not. Clearly this is all dependent upon the **analyst scored dataset** being a decent representation of the population (which it most likely is not). This analysis should be conducted again at some point with a dataset that is more representative of the population. That would require more conversation with subject matter experts, as I do not want to try and manufacture a represenative data set and bring in certain biases.

## 2) Model B and Model D both performed similarly 
**Model B** was a Random Forest trained on the **analyst scored dataset**. It ended up with the following average performance metrics, when making predictions on the **analyst scored dataset**:
* **Accuracy**:  59.8%
* **Precision**: 66.2%
* **Recall**: 76.3%

**Model D** was a Random Forest trained on the **inferred score dataset**. It ended up with the following average performance metrics, when making predictions on the **analyst scored dataset**:
* **Accuracy**:  58.3%
* **Precision**: 86.2%
* **Recall**: 54.2%

What is interesting about this is that since the **inferred score dataset** was not labeled by analysts, it was scored similarly to the rule based algorithm currently in place. When it then made predictions on the domains from the **analyst scored dataset** it performed similarly. This is good and bad. On one hand, this means that the current system is doing a relatively good job, given the predictive quality of the current input features. However, this means that in a real world, ~42% of what it classifies is incorrect. 

## 3) Model B could potentially perform better with more data
Model B was trained on only 1400 labeled examples. If it had been trained on ~ 10,000, it would most likely have an increase in performance.

## 4) Better Input Features are needed
At the end of the day, the current features are not great at predicting whether a domain is valuable or not. Feature engineering will need to be taken advantage of to explain some of the variance in the response. Additionally, further input features such as credit score could be very helpful in this prediction.

## 5) Natural Language Processing is worth exploring
Of the current input features, the only one that has not been experimented with at all is the actual domain name. This feature could be utilized to see if there is any trend, word, phrase, etc, that is correlated to a domain being valuable or not.
